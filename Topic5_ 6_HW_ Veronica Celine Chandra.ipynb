```python
import pandas as pd
import numpy as np
```

```python
#Membuat Data Frame pada soal nomor 1
df1 = pd.DataFrame([["Anna", 45.2, 155, 2000, "Female", "New York"],
                  ["Dane", 50.5, 160, 1997, "Male", "Chicago"],
                  ["David", 62.1, 162, 1996, "Male", "Austin"],
                   ["Kevin", 64.0, 170, 1987, "Male", "New York"],
                   ["Joe", 59.0, 167, 1998, "Male", np.nan],
                  ["Rose", np.nan, 159, 2003, "Female", "New York"]],
                   columns = ["name", "weight", "height", "birth_year", "gender", "city"])
df1
```

```python
df2 = pd.DataFrame([["Anna", "Anna Levis"],
                   ["David", "David John"],
                   ["Dane", "Dane Sebastian"],
                    ["Rose", "Rose Kimberly"]],
                   columns = ["nick_name", "full_name"])
df2
```

```python
#Melakukan merge Data Frame dengan tabel nomor 1 
df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name")
```

```python
#Menghilangkan kolom nick_name pada Data Frame yang telah digabungkan
df1=df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name").drop(columns="nick_name")
df1
```

```python
#Mengganti NaN pada kolom full_name dengan nama pada kolom name
df1['full_name'].fillna(df1['name'],inplace=True)
df1
```

```python
#Membuat kolom baru yaitu kolom BMI menggunakan rumus yang terdapat pada soal nomor 4
df1=df1.replace ('Nan', np.nan)
df1['BMI']=df1['weight']/df1['height'].div(100).pow(2)
df1
```

```python
#Menambahkan kolom BMI_Class sesuai dengan deskripsi yang terdapat pada soal
df1.loc[(df1['BMI']< 18.5), 'BMI_Class'] = "Under weight"
df1.loc[(df1['BMI']>= 18.5)& (df1["BMI"]<25), 'BMI_Class'] = "Normal weight"
df1.loc[(df1['BMI']>= 25)& (df1["BMI"]<30), 'BMI_Class'] = "Over weight"
df1.loc[(df1['BMI']>= 30)& (df1["BMI"]<35), 'BMI_Class'] = "Obesity class 1"
df1.loc[(df1['BMI']>= 35)& (df1["BMI"]<40), 'BMI_Class'] = "Obesity class 2"
df1.loc[(df1['BMI']>= 40), 'BMI_Class'] = "Obesity class 3"
df1
```
